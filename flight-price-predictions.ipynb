## Dataset and attribute description
Dataset contains information about flight booking options from the website Easemytrip for flight travel between India's top 6 metro cities. There are 300261 datapoints and 11 features in the cleaned dataset.

#### Attributes
1. Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
2. Flight: Flight stores information regarding the plane's flight code. It is a categorical feature.
3. Source City: City from which the flight takes off. It is a categorical feature having 6 unique cities.
4. Departure Time: This is a derived categorical feature obtained created by grouping time periods into bins. It stores information about the departure time and have 6 unique time labels.
5. Stops: A categorical feature with 3 distinct values that stores the number of stops between the source and destination cities.
6. Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keeps information about the arrival time.
7. Destination City: City where the flight will land. It is a categorical feature having 6 unique cities.
8. Class: A categorical feature that contains information on seat class; it has two distinct values: Business and Economy.
9. Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours. 
10. Days Left: This is a derived characteristic that is calculated by subtracting the trip date by the booking date.
11. Price: Target variable stores information of the ticket price.

### Question 1: How do prices vary between each class (Business vs Economy)

To visualize the difference in price between Business and Economy, I'll color code each based on the airline.

```python
plt.figure(figsize = (20,5))
sns.barplot(x = 'airline', y = 'price', hue = 'class', data = data.sort_values('price')).set_title('Airline cost based on class and company', fontsize = 20)
```

> Only Air India and Vistara offer Business class flights, at much higher prices than their Economy ticket counterparts. 

### Question 2: Does price vary by airline?

Because some airlines do not offer Business class, I'll only evaluate Economy ticket prices

```python
fig, axs = plt.subplots (1, 2, gridspec_kw={'width_ratios': [5, 3]}, figsize=(25, 5))
sns.violinplot(y = "price", x = "airline",data = data.loc[data["class"]=='Economy'].sort_values("price", ascending = False), kind="boxen", ax=axs[0])
axs[0].set_title("Airline prices based on companies for economy tickets",fontsize=20)
sns.violinplot(y = "price", x = "airline",data = data.loc[data["class"]=='Business'].sort_values("price", ascending = False), kind="boxen", ax=axs[1])
axs[1].set_title("Airline prices based on companies for business tickets",fontsize=20)
```

> As we can see from the plots, AirAsia has the cheapest Economy flights, and Air India has the least expensive Business class flights

### Question 3: How is the price affected when tickets are bought in just 1 or 2 days before departure?

In order to determine how prices change depending on how far or close in advance tickets are purchased, I'll take the average price depending on days left before the flight.

```python
data1 = data.groupby(['days_left'])['price'].mean().reset_index()

plt.figure(figsize = (20,5))
ax = sns.scatterplot(x = 'days_left', y = 'price', data = data1).set_title('Average ticket price by days purchased in advance of flight', 
                                                                           fontsize = 15)
```

> The graph indicates that ticket prices increase as the flight approaches, but the trajectory of that increase changes at the 20-day mark. I'll parse this in the following graph.

```python
plt.figure(figsize=(15,6)).suptitle('Average airline ticket prices based on days left before purchase', fontsize=16)
ax = plt.axes()
sns.regplot(x=data1.loc[data1["days_left"]==1].days_left, y=data1.loc[data1["days_left"]==1].price, fit_reg=False, ax=ax)
sns.regplot(x=data1.loc[(data1["days_left"]>1)&(data1["days_left"]<20)].days_left, y=data1.loc[(data1["days_left"]>1)&(data1["days_left"]<20)].price, fit_reg=True, ax=ax)
sns.regplot(x=data1.loc[data1["days_left"]>=20].days_left, y=data1.loc[data1["days_left"]>=20].price, fit_reg=True, ax=ax)
```

> As we can see, ticket prices slowly rise from 50 days out to 21 days out, and then drastically increase from day 20 to day 1.

### Question 4: How does flight duration affect ticket price?

```python
data1 = data.groupby(['duration'])['price'].mean().reset_index()

plt.figure(figsize = (15,6))
ax = sns.scatterplot(x = 'duration', y = 'price', data = data1).set_title('Flight ticket price by flight duration', fontsize = 15)
```

> Now let's add a best fit line

```python
plt.figure(figsize = (15,6))
ax = sns.scatterplot(x = 'duration', y = 'price', data = data1).set_title('Flight ticket price by flight duration', fontsize = 15)

ax = sns.regplot(x = 'duration', y = 'price', data = data1, order = 2)
```

> The duration-to-price line can be approximated as a quadratic curve, order 2. The curve peaks at around the 20 hour flight time mark.

### Question 5: Does ticket price change based on the departure time and arrival time?

Let's see how departure and arrival times affect prices.

```python
plt.figure(figsize = (20,5))
plt.subplot(1,2,1)
sns.boxplot(x = 'departure_time', y = 'price', data = data, showfliers = False).set_title('Ticket prices based on flight departure time', fontsize=15)
plt.subplot(1,2,2)
sns.boxplot(x= 'arrival_time', y= 'price', data = data, showfliers = False).set_title('Ticket prices based on flight arrival time', fontsize = 15)
```

> As we can see, early morning and late night ticket prices tend to be the least expensive for arrivals, while night and evening departures tend to be the most expensive for departures and arrivals.

### Question 6: Does ticket price change based on the departure and arrival cities?

```python
ax = sns.relplot(data = data, col = 'source_city', x = 'destination_city', y = 'price', kind = 'line', col_wrap = 3)
ax.fig.subplots_adjust(top = .9)
ax.fig.suptitle('Ticket prices based on Arrival and Departure City', fontsize = 20)
```

> From the charts, we can see that flights are least expensive out of Dehli, while most expensive out of Bangalore.

### Question 7: Does the number of stops influence the price?

```python
fig, axs = plt.subplots(1,2, gridspec_kw = {'width_ratios' : [5,3]}, figsize = (20,5))
sns.barplot(data = data.loc[data['class']=='Economy'].sort_values('price', ascending = False), x = 'airline', y = 'price', hue = 'stops', ax = axs[0])
axs[0].set_title('Economy class tickets based on number of stops', fontsize = 15)
sns.barplot(data = data.loc[data['class']=='Business'].sort_values('price', ascending = False), x = 'airline', y = 'price', hue = 'stops', ax = axs[1])
axs[1].set_title('Business class tickets based on number of stops', fontsize = 15)
```

> For each airline, more stops tends to translate to higher ticket prices, while non-stop flights tend to be least expensive. For AirAsia, though, ticket prices tend to be constant regardless of number of stops.

### Key insights from EDA
* Business class flights are more expensive than economy class flights, and a major predictor of ticket price
* Flights around the 20-hr mark are the peak of expensive flights, as flights longer than 20 hours decrease in price.
* Flights become more expensive as the day of flight approaches
* Late night departures and arrivals are least expensive, and night departures are most expensive

## Price Predictions

```python
final_data = pd.read_csv('https://raw.githubusercontent.com/julienjta/Data-Science-Portfolio/main/Project%201/data/dataset.csv',index_col=0)
```

```python
final_data.head()
```

##### Importing packages

```python
from sklearn.model_selection import KFold, cross_val_predict, GridSearchCV, train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, LabelEncoder
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV, ElasticNetCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
```

```python
for i in final_data.select_dtypes('object').columns:
    encoder = LabelEncoder()
    final_data[i] = encoder.fit_transform(final_data[i])
```

```python
final_data.drop('flight', axis=1, inplace=True)
```

```python
final_data
```

##### Train/test split

```python
X = final_data.drop('price', axis = 1)
y = final_data.price
```

```python
kf = KFold(shuffle=True, random_state=42, n_splits=3)
```

```python
kf.split(X)
```

```python
X.shape
```

### Simple Linear model

```python
s = StandardScaler()
lr = LinearRegression()
estimator = Pipeline([('scaler', s), ('regression', lr)])
predictions = cross_val_predict(estimator, X, y, cv = kf)
```

```python
linear_r2 = r2_score(y, predictions)
print('R2 score of polynomial linear regression is', linear_r2)
```

### Linear model with polynomial features

```python
pf = PolynomialFeatures()
estimator2 = Pipeline([('poly', pf), ('scaler', s), ('regression', lr)])
parameters = {'poly__degree': np.arange(2,5)}

grid1 = GridSearchCV(estimator2, parameters, cv = kf)
```

```python
grid1.fit(X, y)
```

```python
print('Optimal degrees of polynomial linear regression is', grid1.best_params_['poly__degree'])

estimator2 = grid1.best_estimator_
y_pred = estimator2.predict(X)

poly_r2 = r2_score(y, y_pred)

print('R2 score of polynomial linear regression is', poly_r2)
```

```python
from matplotlib.pyplot import xlabel


f = plt.figure(figsize= (12,12))
ax = plt.axes()

ax.plot(y, y_pred, marker = 'o', ls = '')

lim = (0, y_pred.max())

ax.set(xlabel = 'Actual Price',ylabel = 'Predicted Price', title = 'Polynomial regression results')

ax = sns.regplot(x = y, y = y_pred, data= final_data, order = 4).set_title('Actual prices vs predicted prices', fontsize = 20)
```

### Key insights from model predictions
